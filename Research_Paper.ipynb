# **LLMs in Job Recruitment: Can AI Replace the First Interview?**

---

### **Introduction**

In today's competitive job market, employers face an unprecedented challenge in recruiting top talent in an effective and fair way. The global rise in remote work, an increasing emphasis on diversity, equity, and inclusion (DEI), and the desire to reduce hiring costs have accelerated the quest for scalable hiring solutions. Consequently, businesses are turning to Artificial Intelligence (AI), and more precisely Large Language Models (LLMs), to automate initial stages of the hiring process. LLMs like OpenAI’s ChatGPT-4 and Google’s Gemini claim to replicate human-like conversational abilities, raising the question: can they replace human recruiters in conducting first-round interviews?

This research paper explores whether LLMs are able to adequately perform early-stage interviews by experimentation in simulated hiring scenarios. It also examines the overall social, ethical, and technological implications of using AI for a task so intensely interpersonal and judgmental. While AI brings efficiency, cost savings, and consistency, its shortcomings in emotional intelligence, fairness, and minimization of bias may challenge its role as a total replacement of human judgment in the hiring process.

---

### **Historical Context: Automation in Hiring**

Automating the hiring process is not an innovative idea. During the 1990s, businesses started to apply Applicant Tracking Systems (ATS) to filter resumes based on keyword matches. The systems valued efficiency above other concerns but were highly criticized for eliminating competent applicants who had not strategically formatted their resumes (Cappelli, 2001).

With the advent of machine learning, these technologies were more sophisticated, employing sentiment analysis and predictive analytics to assess candidate fit. Towards the late 2010s, video interview analysis technology such as HireVue purported to analyze facial expressions and tone of voice to assess soft skills and cultural fit. These technologies were attacked, however, after a study revealed their propensity to be infused with racial and gender biases (Ajunwa, 2020).

LLMs represent the next generation in AI hiring—capable not only to screen but also to communicate. They have created new controversy regarding fairness, transparency, and the very nature of human resource management.

---

### **Literature Review: AI’s Capabilities and Ethical Risks in Recruitment**

#### **Technical Ability of LLMs**

LLMs are proficient in language understanding and generation, having been trained on massive text material datasets. This enables them to simulate human chat and ask relevant follow-up questions based on candidate responses. However, as Marcus and Davis (2022) note, these models remain essentially pattern-matching software, not agents that can carry out real reason or judgment.

Contemporary LLMs lack genuine theory of mind—the sort of mental capacity to read other individuals' mental states. While they can mimic empathetic speech, they are not self-conscious and lack genuine emotional comprehension, so it is difficult for them to assess complex interpersonal relations or detect insincerity—both of which are essential interviewing skills.

#### **Bias and Discrimination**

Algorithmic hiring has already shown real advantages. Amazon's failed AI hiring tool is an example; biased on historical hiring data by male applicants for technical positions, the program punished resumes that included women's groups (Dastin, 2018).

While companies today aim to offset such biases, the problem runs deep. As Raji et al. (2020) explain, AI models are likely to inherit social biases present in their training data and reproduce them at scale and with the aura of objectivity. The embedding of LLMs into hiring pipelines might further entrench such problems unless companies audit and retrain their models on a regular basis.

#### **Transparency and Accountability**

Raghavan et al. (2020) point to the possibility of enhanced accountability gap with AI-driven hiring. When a human recruiter chooses to hire, his logic can be queried; when an AI does so, it is usually the result of some advanced and inscrutable model architecture. That lack of transparency poses fundamental concerns regarding applicants' rights to appeal or dispute hiring decisions.

The EEOC (2023) has issued guidelines warning employers against potential anti-discrimination law violations in the application of algorithmic decision-making tools, but enforcement remains limited.

### **Case Study: Simulating First-Round Interviews with LLMs**

#### **Methodology**

To examine the effectiveness of LLMs in real-world hiring scenarios, I designed a simulation involving two advanced models—ChatGPT-4 and Gemini. Both were tasked with interviewing a simulated candidate for a Product Management Intern position. The interview focused on key competencies: leadership potential, communication skills, problem-solving ability, and familiarity with agile workflows.

Each LLM was prompted with identical background information and evaluated on conversational adaptability, depth of inquiry, emotional sensitivity, and final hiring recommendations.

---

#### **Prompt/Response Chain**

**Scenario Context:**
“You are a recruiter conducting a first-round interview for a Product Management Intern. Assess the candidate’s leadership potential, communication skills, and problem-solving ability. Ask open-ended questions, adapt follow-ups based on responses, and conclude with a hiring recommendation.”

---

**Candidate Simulation:**
“Hi, I’m Alex Chen. Thanks for speaking with me today!”

**ChatGPT-4 Response:**
“Hi Alex, it’s great to meet you! Let’s start by talking about leadership. Can you share a time when you led a project or initiative? What challenges did you face, and how did you overcome them?”


**Candidate:**
“In my sophomore year, I led a team of five students to develop a mobile app connecting students with local volunteer opportunities. The biggest challenge was keeping everyone engaged when academic workloads increased. I organized weekly stand-ups and created a shared progress board to track progress.”


**ChatGPT-4 Follow-Up:**
“That’s a creative approach! How did you measure the success of this initiative, and reflecting on the experience, what would you do differently next time?”


**Candidate:**
“We had over 300 signups in the first month. Looking back, I would have incorporated user feedback earlier in the design process to improve usability.”


**ChatGPT-4 Final Internal Note:**
*Strong leadership and communication skills demonstrated. Candidate displays critical thinking and growth mindset. Recommend advancing to the next stage.*

---

**Gemini Response:**
“Hello Alex. Could you describe a situation where you worked under pressure and how you managed it?”

*Gemini moved directly into scripted behavioral questions without responding to or building upon previous answers, leading to a less dynamic interview flow.*

---

**Observations:**
ChatGPT-4 delivered a more adaptive and engaging interview, responding fluidly to candidate answers and probing deeper into project outcomes. Gemini adhered rigidly to a question script, missing opportunities for deeper insight.

Neither model addressed long-term career goals or cultural fit, elements typically explored by human interviewers to assess organizational alignment. Additionally, both models struggled to ask emotionally nuanced questions, such as how the candidate handled interpersonal conflict—critical for leadership evaluation.

---

### **Deeper Ethical Analysis: What Are We Losing in Human-Less Hiring?**

The primary danger of AI interviews is not just skewed outcomes but the insidious erosion of human-centric values in recruitment. Recruitment is relationally grounded, contends McStay (2023), and this involves assessing emotional resonance, cultural fit, and an oft-unspoken intuition of "potential" that cannot be quantified by pattern recognition.

Moreover, dependence on LLMs risks commodifying job applicants. Converting a rich human into a set of behavioral responses that an algorithm can break down reduces the richness of human experience and individuality. This dehumanization can pay long-term dividends in job contentment, motivation, and employee turnover, as candidates screened by machine might not be optimally placed into workable positions within a company's actual culture and values.

---

### **Future Outlook: Should LLMs Be Gatekeepers to Employment?**

In the future, the use of AI in hiring will expand. LinkedIn and Indeed already employ AI tools to recommend job matches and evaluate candidate fit. Startups are developing AI interviewers trained for various industries, and some companies even see a future where candidates only interact with AI throughout the hiring process.

But these trends must be heeded. LLMs cannot serve as gatekeepers to possibility except under rigorous regulation, ethical protection, and controls for human oversight. The best path is to integrate LLMs as support tools—simplifying mechanical administrative procedures without eradicating the human element in determinations that set people's careers and livelihoods. 

--- 

### **Conclusion**

LLMs like ChatGPT-4 and Gemini have strong potential to conduct structured interviews and adapt surface-level skills. They fall short in assessing deeper soft skills, emotional intelligence, and authenticity. While they can aid recruiters by handling huge volumes of applicants and reducing initial screening burdens, they cannot—and must not—replace human judgment in assessing the full spectrum of human potential.

To be equitable, companies must prioritize transparency, provide candidates with the option to request human review, and regularly audit their AI hiring tools for bias. The future of recruitment must be one in which AI augments and assists human judgment, preserving the dignity, fairness, and empathy demanded of good hiring practices.
